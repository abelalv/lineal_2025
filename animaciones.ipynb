```python
# geracion de un git para  ver que ver que la la combinacion lineal de 
# dos vectores generan r^2 

#| echo: false
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import Image, display

v1 = np.array([2, 1])
v2 = np.array([1, 2])

# Configurar la figura
fig, ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-30, 30)
ax.set_ylim(-30, 30)
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Generación de cualquier punto en el espacio')

def update(frame):
  ax.cla()  # Limpiar ejes
  ax.set_xlim(-30,30)
  ax.set_ylim(-30,30)
  ax.axhline(0, color='gray', linewidth=0.5)
  ax.axvline(0, color='gray', linewidth=0.5)
  ax.set_xlabel('Eje X')
  ax.set_ylabel('Eje Y')
  ax.set_title('Generación de cualquier punto en el espacio')
  
  # Reducir el factor de crecimiento para que los vectores permanezcan en el espacio.
  t = frame / frames
  r = 10 * t                # Valor máximo reducido para evitar que los vectores se salgan
  theta = 4 * np.pi * t     # Dos vueltas completas
  c1 = r * np.cos(theta)
  c2 = r * np.sin(theta)
  
  vec = c1 * v1 + c2 * v2
  vec1 = c1 * v1
  vec2 = c2 * v2
  
  # Dibujar el vector resultante y sus componentes
  ax.quiver(0, 0, vec[0], vec[1],
        color='red', angles='xy', scale_units='xy', scale=1,
        clip_on=True, label=f"$({c1:.2f})v_1+({c2:.2f})v_2$")
  ax.quiver(0, 0, vec1[0], vec1[1],
        color='blue', angles='xy', scale_units='xy', scale=1,
        clip_on=True, label=f"$({c1:.2f})v_1$")
  ax.quiver(vec1[0], vec1[1], vec2[0], vec2[1],
        color='green', angles='xy', scale_units='xy', scale=1,
        clip_on=True, label=f"$({c2:.2f})v_2$")
  
  ax.legend()
  ax.grid(True)

frames = 60
# Aumentamos el intervalo para que la animación sea más lenta (300 ms entre fotogramas)
anim = FuncAnimation(fig, update, frames=np.linspace(0, frames, frames), interval=300, blit=False)
gif_filename = "animation_spiral.gif"
anim.save(gif_filename, writer="imagemagick")
plt.close()
display(Image(filename=gif_filename))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

v = np.array([1, 2])

def plot_span(alpha):
  plt.figure(figsize=(6,6))
  # Rango para la línea que representa el span de v
  t = np.linspace(-10, 10, 400)
  points = np.outer(t, v)
  plt.plot(points[:, 0], points[:, 1], label='Span de v', color='black')
  
  # Vector escalado según el valor de alpha
  vector_actual = alpha * v
  plt.quiver(0, 0, vector_actual[0], vector_actual[1], angles='xy', scale_units='xy', scale=1,
             color='red', label=f"${alpha}\\cdot v$")
  
  plt.axhline(0, color='gray', linewidth=0.5)
  plt.axvline(0, color='gray', linewidth=0.5)
  plt.xlabel('Eje X')
  plt.ylabel('Eje Y')
  plt.legend()
  plt.grid(True)
  plt.title('Span de un solo vector con slider')
  plt.show()

from ipywidgets import FloatSlider
interact(plot_span, alpha=FloatSlider(min=-10, max=10, step=0.5, value=1, description="c"))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

v = np.array([1, 2])

def plot_span(alpha):
  plt.figure(figsize=(6,6))
  # Rango para la línea que representa el span de v
  t = np.linspace(-10, 10, 400)
  points = np.outer(t, v)
  plt.plot(points[:, 0], points[:, 1], label='Span de v', color='black')
  
  # Vector escalado según el valor de alpha
  vector_actual = alpha * v
  plt.quiver(0, 0, vector_actual[0], vector_actual[1], angles='xy', scale_units='xy', scale=1,
             color='red', label=f"${alpha}\\cdot v$")
  
  plt.axhline(0, color='gray', linewidth=0.5)
  plt.axvline(0, color='gray', linewidth=0.5)
  plt.xlabel('Eje X')
  plt.ylabel('Eje Y')
  plt.legend()
  plt.grid(True)
  plt.title('Span de un solo vector con slider')
  plt.show()

from ipywidgets import FloatSlider
interact(plot_span, alpha=FloatSlider(min=-10, max=10, step=0.5, value=1, description="c"))
# Crear una animación del span de v
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-30, 30)
ax.set_ylim(-30, 30)
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Animación del span de v')

def update(frame):
    ax.cla()
    ax.set_xlim(-30, 30)
    ax.set_ylim(-30, 30)
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    ax.set_xlabel('Eje X')
    ax.set_ylabel('Eje Y')
    ax.set_title('Animación del span de v')
    
    # Rango para la línea que representa el span de v
    t = np.linspace(-10, 10, 400)
    points = np.outer(t, v)
    ax.plot(points[:, 0], points[:, 1], label='Span de v', color='black')
    
    # Vector escalado según el valor de frame
    alpha = frame / 10 - 10
    vector_actual = alpha * v
    ax.quiver(0, 0, vector_actual[0], vector_actual[1], angles='xy', scale_units='xy', scale=1,
              color='red', label=f"${alpha:.2f}\\cdot v$")
    
    ax.legend()
    ax.grid(True)

frames = np.linspace(0, 200, 200)
anim = FuncAnimation(fig, update, frames=frames, interval=100, blit=False)
gif_filename = "span_animation.gif"
anim.save(gif_filename, writer="imagemagick")
plt.close()
```
